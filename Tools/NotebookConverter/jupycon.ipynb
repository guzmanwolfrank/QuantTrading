converts from jupyter notebook to python .py file


```python
import nbformat
from nbconvert import PythonExporter

# Define the paths
input_path = r'C:\Users\wolfr\OneDrive\Desktop\algohausBTE.ipynb'
output_path = r'C:\Users\wolfr\OneDrive\Desktop\algohausBTE.py'

# Read the notebook with UTF-8 encoding
with open(input_path, 'r', encoding='utf-8') as f:
    notebook = nbformat.read(f, as_version=4)

# Convert to Python
exporter = PythonExporter()
python_code, _ = exporter.from_notebook_node(notebook)

# Save as .py file with UTF-8 encoding
with open(output_path, 'w', encoding='utf-8') as f:
    f.write(python_code)

print(f"✓ Conversion complete!")
print(f"✓ File saved to: {output_path}")
```
